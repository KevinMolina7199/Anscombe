# Análisis de Datos de Anscombe

Este documento describe el análisis de los datos del conjunto de datos de Anscombe, que es conocido por mostrar la importancia de la visualización de datos. El objetivo es analizar diferentes grupos dentro del conjunto de datos y visualizar las relaciones entre las variables.

## Importar Librerías Necesarias

Se importan las bibliotecas necesarias para el análisis y visualización de datos.



```python
# Importar librerías necesarias
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

## Cargar el Dataset
Se carga el conjunto de datos de Anscombe desde un archivo CSV.

```python
ruta='../data/df_anscombe.csv'
dataset=pd.read_csv(ruta)
```

# Resumen Estadístico por Grupo
Se realiza un resumen estadístico agrupando los datos por la columna 'group'. Esto permite observar estadísticas descriptivas como la media, la desviación estándar, etc.

```python
dataset.groupby('group').describe()
```

```python
dataset.info()
```

# Gráficos de Cajas y Bigotes
Se crean gráficos de cajas y bigotes para visualizar la distribución de las variables 'x' y 'y' agrupadas por 'group'.

```python
# Configurar el tamaño de los gráficos
plt.figure(figsize=(10, 6))

# Crear el gráfico de cajas y bigotes para la variable 'x' agrupado por 'group'
sns.boxplot(x='group', y='x', data=dataset)
plt.title('Gráfico de Cajas y Bigotes de x por Grupo')
plt.show()

# Crear el gráfico de cajas y bigotes para la variable 'y' agrupado por 'group'
plt.figure(figsize=(10, 6))
sns.boxplot(x='group', y='y', data=dataset)
plt.title('Gráfico de Cajas y Bigotes de y por Grupo')
plt.show()
```

# Gráficos de Dispersión por Grupo
Se crean gráficos de dispersión para cada grupo para observar la relación entre 'x' e 'y'.

```python
# Configurar el tamaño de los gráficos
plt.figure(figsize=(10, 6))

# Filtrar por grupo 1 y crear el gráfico
plt.subplot(2, 2, 1)
sns.scatterplot(x='x', y='y', data=dataset[dataset['group'] == 1], color='blue', s=100)
plt.title('Grupo 1')
plt.xlabel('x')
plt.ylabel('y')

# Filtrar por grupo 2 y crear el gráfico
plt.subplot(2, 2, 2)
sns.scatterplot(x='x', y='y', data=dataset[dataset['group'] == 2], color='green', s=100)
plt.title('Grupo 2')
plt.xlabel('x')
plt.ylabel('y')

# Filtrar por grupo 3 y crear el gráfico
plt.subplot(2, 2, 3)
sns.scatterplot(x='x', y='y', data=dataset[dataset['group'] == 3], color='red', s=100)
plt.title('Grupo 3')
plt.xlabel('x')
plt.ylabel('y')

# Filtrar por grupo 4 y crear el gráfico
plt.subplot(2, 2, 4)
sns.scatterplot(x='x', y='y', data=dataset[dataset['group'] == 4], color='orange', s=100)
plt.title('Grupo 4')
plt.xlabel('x')
plt.ylabel('y')

# Ajustar el diseño para que los gráficos no se solapen
plt.tight_layout()

# Mostrar los gráficos
plt.show()
```

# Gráficos de Regresión Lineal por Grupo
Se crean gráficos de regresión lineal para cada grupo para observar la tendencia entre 'x' e 'y'.

```python
# Configurar el tamaño de los gráficos
plt.figure(figsize=(10, 6))

# Grupo 1
plt.subplot(2, 2, 1)
sns.regplot(x='x', y='y', data=dataset[dataset['group'] == 1], color='blue', ci=None)
plt.title('Grupo 1 - Dispersión y Regresión Lineal')
plt.xlabel('x')
plt.ylabel('y')

# Grupo 2
plt.subplot(2, 2, 2)
sns.regplot(x='x', y='y', data=dataset[dataset['group'] == 2], color='green', ci=None)
plt.title('Grupo 2 - Dispersión y Regresión Lineal')
plt.xlabel('x')
plt.ylabel('y')

# Grupo 3
plt.subplot(2, 2, 3)
sns.regplot(x='x', y='y', data=dataset[dataset['group'] == 3], color='red', ci=None)
plt.title('Grupo 3 - Dispersión y Regresión Lineal')
plt.xlabel('x')
plt.ylabel('y')

# Grupo 4
plt.subplot(2, 2, 4)
sns.regplot(x='x', y='y', data=dataset[dataset['group'] == 4], color='orange', ci=None)
plt.title('Grupo 4 - Dispersión y Regresión Lineal')
plt.xlabel('x')
plt.ylabel('y')

# Ajustar el diseño para que los gráficos no se solapen
plt.tight_layout()

# Mostrar los gráficos
plt.show()
```

# Gráfico de Dispersión General
Se crea un gráfico de dispersión que muestra todos los grupos en un solo gráfico, con diferentes colores para cada grupo.

```python
# Configurar el tamaño del gráfico
plt.figure(figsize=(10, 6))

# Gráfico de dispersión con seaborn, coloreando por 'group'
sns.scatterplot(x='x', y='y', hue='group', data=dataset, palette='Set1', s=100)

# Añadir título y etiquetas
plt.title('Gráfico de Dispersión de x vs y agrupado por Group')
plt.xlabel('x')
plt.ylabel('y')

# Mostrar el gráfico
plt.show()
```

# Conclusiones
A través de este análisis, se pueden observar las diferencias en la relación entre 'x' e 'y' para cada grupo, así como la variabilidad en las distribuciones. Los gráficos de cajas y dispersión ayudan a identificar tendencias y posibles outliers en los datos.
